# Gradient Boosting with XGBoost

## Introduction
Nella sezione **Gradient Boosting**, si imparerà a conoscere XGBoost, una libreria per alberi decisionali gradient boosted altamente efficienti. È una delle principali librerie utilizzate in data science per la classificazione e la regressione.

## XGBoost vs. scikit-learn
Le precedenti 3 sezioni di questo corso hanno utilizzato scikit-learn per una serie di tasks. 
`XGBoost` è una libreria di data science all'avanguardia per l'esecuzione di classificazioni e regressioni. 
- XGBoost fa uso di alberi decisionali potenziati dal gradiente, che forniscono prestazioni migliori rispetto ai normali alberi decisionali.

Oltre all'aumento delle prestazioni, XGBoost implementa una versione estremamente efficiente degli alberi a gradiente potenziato. 
- I modelli XGBoost si **addestrano molto più velocemente** dei modelli scikit-learn, pur garantendo la stessa facilità d'uso.

Nelle competizioni di data science e di machine learning che utilizzano insiemi di dati di piccole e medie dimensioni (ad esempio Kaggle), XGBoost è sempre tra i modelli più performanti.

## Gradient boosted trees
Il **problema** dei normali alberi decisionali è che spesso non sono abbastanza complessi per catturare le complessità di molti set di dati di grandi dimensioni. Potremmo aumentare continuamente la profondità massima di un albero decisionale per adattarlo a insiemi di dati più grandi, ma gli alberi decisionali con molti nodi tendono ad adattarsi troppo ai dati.
- Invece, utilizziamo il `gradient boosting` per combinare molti alberi decisionali in un unico modello per la classificazione o la regressione. 
- Il gradient boosting inizia con un singolo albero decisionale, 
- quindi aggiunge iterativamente altri alberi decisionali al modello complessivo per correggere gli errori del modello sul set di dati di allenamento.

L'API XGBoost gestisce il processo di gradient boosting per noi, producendo un modello molto migliore di quello che avremmo ottenuto utilizzando un singolo albero decisionale.

### Time to Code!
L'esercizio consiste nell'addestrare un oggetto `Booster` sui `data` di input e sulle `labels` (predefinite nel backend).
- La prima cosa da fare è impostare `DMatrix` per l'addestramento.
- Setta `dtrain` uguale a `xgb.DMatrix` inizializzato con `data` come argomento richiesto e `labels` come argomento della keyword `label`.

```python
dtrain = xgb.DMatrix(data, label=labels)
```

Il set di dati di input contiene 3 classi, quindi si eseguirà una classificazione multiclasse con il `Booster`. L'insieme di dati è anche relativamente piccolo, quindi la profondità massima dell'albero decisionale è limitata a 2.
- Ciò significa che i parametri dell'oggetto `Booster` avranno `'max_depth'` settato a `2`, `'objective'` settato a `'multi:softmax'` e `'num_class'` settato a 3.
- Setta `params` come un dizionario con le chiavi e i valori specificati.

```python
params = {
    'max_depth': 2,
    'objective':'multi:softmax',
    'num_class': 3
}
```

Utilizzando la matrice di dati e i parametri, si addestri il Booster.
- Setta `bst` uguale a `xgb.train` applicato con `params` e `dtrain` come argomenti richiesti.

```python
bst = xgb.train(params, dtrain)
```
