## Linear Regression
### What is linear regression?
Uno dei principali obiettivi, sia nel machine learning che in data science, è trovare un equazione o distribuzione che si adatti al meglio a un dato dataset. Questo è conosciuto come `data modelling`, dove si creare un modello che usi le caratteristiche del dataset come variabile indipendenti per prevedere i valori di output per alcune variabili dipendenti (con un errore minimo).
E' incredibilemente difficile trovare un modello ottimale per la maggior parte dei dataset, data la quantità di rumore (cioè errori/fluttuazioni casuali) nei dati del mondo reale.
<br><br>Poichè trovare un modello ottimale per un dataset è difficile, invece si cerchi di trovare una buona distribuzione approssimativa. 
In molti casi, un modello lineare (una combinazione lineare delle caratteristiche del set di dati) può approssimare bene i dati.
<br><br>Il termine `regressione lineare` si riferisce all'utilizzo di un modello lineare per rappresentare la relazione tra un 'insieme di variabili indipendenti e una variabile dipendente.

```python
Image("regressione_lineare.png")
```

La formula sopra è un esempio di modello lineare che produce output `y` (variabile dipendente) basata su una combinazione lineare di variabili indipendenti `x_1`, `x_2`,`x_3`. I coefficienti `a`, `b`, `c` e intercetta `d`, determinano l'andamento del modello.

### Basic linear regression
La più semplice forma di regressione lineare è chiamata `least squares regression`. Questa strategia pridce una modello di regressione, che è una combinazione lineare di variabili indipendenti, che minimizzano la `somma dei quadrati residui` tra le stime del modello e i valori effettivi per la variabile dipendente.
<br><br>In scikit-learn, il modello di regressione dei minimi quadrati è implementato con l'oggetto `LinearRegression`, che è parte del modulo `linear_model` in `sklearn`. L'oggetto contiene la funzione `fit`, che accetta un set di dati di input di funzionalità (variabili indipendenti) e una matrice di etichette (variabili dipendenti) per ogni osservazione dei dati (righe del set di dati).

```python
# prima colonna di pizza_data: calorie
# seconda colonna di pizza_data: grammi

pizza_data = np.array([[2100,  800],
                       [2500,  850],
                       [1800,  760],
                       [2000,  800],
                       [2300,  810]])
pizza_prices = np.array([10.99, 12.5 ,  9.99, 10.99, 11.99])

# predefined pizza data and prices
print('{}\n'.format(repr(pizza_data)))
print('{}\n'.format(repr(pizza_prices)))
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(pizza_data, pizza_prices)
```

Dopo aver chiamato la funzione `fit`, il modello è pronto per essere usato. La funzione `predict` ci permette di fare predizione su nuovi dati.
<br><br>Si possno anche ottenere i coefficienti specifici e l'intercetta per la combianzione lineare usando rispettivamente le proprietà `coef_` e `intercept_`.
<br><br>Infine si può recuperare il `coefficiente di determinazione` (o il valore R_quadro) usando la funzione `score` applicata aL dataset e alle etichette. Il valore 
`R_quadro` dice quanto è vicino di un adattamento il modello lineare ai dati, o in altre parole, quanto è buono un adattamento del modello per i dati.

```python
# new pizza data
new_pizzas = np.array([[2000,  820],
                       [2200,  830]])
```

```python
price_predicts = reg.predict(new_pizzas)
print('{}\n'.format(repr(price_predicts)))

print('Coefficients: {}\n'.format(repr(reg.coef_)))
print('Intercept: {}\n'.format(reg.intercept_))
```

```python
# Using previously defined pizza_data, pizza_prices
r2 = reg.score(pizza_data, pizza_prices)
print('R2: {}\n'.format(r2))
```

Il valore `R quadro` tradizionale è un numero reale compreso tra 0 e 1. 
- In scikit-learn va da -∞ a 1, dove valori più bassi denotano un modello più povero adatto ai dati. 
- Più il valore è vicino a 1, migliore è l'adattamento del modello ai dati. 

Nell'esempio sopra, si vede che il modello è quasi perfetto per i dati della pizza

### Time to Code!
L'esercizio usa l'oggetto `LinearRegression` del modulo `linear_model` per completare la funzione `linear_reg`.
- La funzione adatterà un modello di regressione dei minimi quadrati di base ai dati di input e alle etichette.
<br><br>
- setta `reg` uguale a `linear_model.LinearRegression` inizializzato con nessun imput di argoemento
- Chiama `reg.fit` con `data` e `labels` come 2 input di argomenti. Ritorna `reg`.

```python
def linear_reg(data, labels):
    reg = linear_model.LinearRegression()
    reg.fit(data, labels)
    return reg
```

## Ridge Regression
Mentre la regressione dei `minimi quadrati ordinari` è un buon modo per adattare un modello lineare a un set di dati, la `ridge regression` si basa sul fatto che le caratteristiche del set di dati sono indipendenti, cioè non correlate. Quando molto delle funzionalità del set di dati sono correlate linearmente,
- ad esempio, se un set di dati ha più funzionalità che rappresentano lo stesso prezzo in diverse valute

rende il modello di regressione dei minimi quadrati altamente sensibile al rumore dei dati.
<br><br>
Poichè i dati della vita reale tendono ad avere rumore e spesso hanno alcune carattestiche linearmente correlate nel set di dati, si affronta questo problema eseguendo la `regolarizzazione`. 
- Per la regressione dei minimi quadrati ordinari, l'obiettivo è trovare i pesi (coefficienti) per il modello lineare che minimizzino la somma dei residui al quadrato:

```python
Image("regolarizzazione.png")
```

dove ogni `xi` rappresenta un'osservazione di dati e `yi` è la corrispondente etichetta.
<br><br>Per la regolarizzazione, l'obiettivo non è solo quello di minimizzare la somma dei quadrati residui, ma farlo con i coefficienti il più piccoli possibile. 
- Più piccoli sono i coefficienti, meno sono suscettibili al rumore casuale nei dati. La forma più comunemente usata di regolarizzazione è la ridge regularization.
<br><br> Con la `ridge regularization`, l'obiettivo è ora quello di trovare i pesi che riducono al minimo la seguente quantità:

```python
Image("ridge_regularization.png")
```

dove `α` è un iperparametro di un numero reale non negativo e `||w||2` rappresenta la norma `L2` dei pesi. L'addizionale

```python
Image("ridge_regularization1.png")
```

E' indicato come il termine di panlità, poichè penalizza i valori di peso maggiori. Maggiori quantità di α porranno maggiore enfasi sul termine di penalità, constringendo il modello ad avere valori di peso ancora più piccoli.

```python
Image("ordinary_least_square_vs_ridge.png")
```

Il grafico sopra mostra un esmepio di modelli di regressione dei minimi quadrati ordinari rispetto ai modelli di regressione della cresta. Le due croci rosse segnano i punti (0.5, 0.5) e (1, 1), e le linee blu sono le linee di regresione per questi due punti.<br>Ciascuna delle linee grigie sono le linee di regressione per i punti originali con rumore aggiunto (che è indicato dai punti grigi).
<br><br>La regressione dei minimi quadrati ordinari è molto più suscettibile di essere influenzata dal rumore aggiunto, poiché esiste un grado molto maggiore di varianza nelle linee di regressione grigie rispetto alla regressione della cresta.

### Choosing the best alpha
In scikit-learn, si implementa la `ridge regression` essenzialmente nello stesso modo in cui si implementa la `ordinary least squares regression`. Si usa l'oggetto `Ridge` (parte del modulo `linear_model`) per implementare la ridge regression.

```python
from sklearn import linear_model
reg = linear_model.Ridge(alpha=0.1)
reg.fit(pizza_data, pizza_prices)
```

```python
print('Coefficients: {}\n'.format(repr(reg.coef_)))
print('Intercept: {}\n'.format(reg.intercept_))
```

```python
r2 = reg.score(pizza_data, pizza_prices)
print('R2: {}\n'.format(r2))
```

Si può specificare il valore dell'iperparametro α quando si inizializza l'oggetto `Ridge` (il valore di default è 1.0). Tuttavia, invece di scegliere manualmente un valore, si può utizzare la cross-validation per sceglire α ottimale da un elenco di valori.
<br><br>Si parlerà nello specifico della cross-validation nei capitoli Cross-validation, Applying CV to Decision Trees e Exhaustive Tuning, ma per ora basti sapere che si può implementare una `ridge regression` con cross-validation usando l'oggetto `Ridge`.

```python
from sklearn import linear_model
alphas = [0.1, 0.2, 0.3]
reg = linear_model.RidgeCV(alphas=alphas)
reg.fit(pizza_data, pizza_prices)
```

```python
print('Coefficients: {}\n'.format(repr(reg.coef_)))
print('Intercept: {}\n'.format(reg.intercept_))
print('Chosen alpha: {}\n'.format(reg.alpha_))
```

### Time to Code!
Si usa l'oggetto `RidgeCV` del modulo `linear_model` per completare la funzione `cv_ridge_reg`.
La funzione adatterà un modello a ridge regression a un input di dati ed etichette. Il modello è cross-validated per scegliere al meglio il valore di α dalla lista in input, `alphas`.
- Setta `reg` uguale a `linear_model.RidgeCV` inizializzato con un lista di input, `alphas`, per l'argomento della keyword `alphas`.
- Chiama `reg.fit` con `data` e `labels` come 2 argomenti di input. Si ritorni `reg`.

```python
def cv_ridge_reg(data, labels, alphas):
    reg = linear_model.RidgeCV(alphas=alphas)
    reg.fit(data, labels)
    return reg
```

```python
Image("lassoregolarization.png")
```

## Bayesian Regression
### Bayesian techniques
Finora si è discusso dell'ottimizzazione degli iperparametri attraverso la **cross-validation**. Un altro modo per ottimizzare gli iperparametri di un modello di regressione regolarizzato è con **tecniche bayesiane**.
<br><br>
Nella statistica bayesiana, l'idea principale è quella di fare alcunne ipotesi sulle distribuzioni di probabilità dei parametri di un modello prima di essere montati sui dati. Queste ipotesi di distribuzione iniziali sono chiamate **priori** per i parametri del modello.
<br><br>
In un modello di regressione della cretsa bayesiana, ci sono due iperparametri da ottimizzare: α e λ. 
- L'iperparametro `α` ha lo stesso identico scopo della regressione regolare della cresta; vale a dire, funge da **fattore di scala** per il **termine di penalità**.
- L'iperparametro `λ` agisce come la **precisione dei pesi del modello**. Fondamentalmente, minore è il valore λ, maggiore è la varianza tra i singoli valori di peso.

### Hyperparameter priors
Sia l'iperparametro α che λ hanno **priorità di distribuzione gamma**, il che significa che si assuma che entrambi i valori provengano da una distribuzione di probabilità gamma.
<br><br>
Non è necessario conoscere le specifiche di una distribuzione gamma, a parte il fatto che è una **distribuzione di probabilità definita** da un parametro di `forma` e di `scala`.
- l'iperaparametro `α` ha priorità

```python
Image("iperparametro_alfa.png")
```

- e l'iperparametro `λ` ha priorità:

```python
Image("iperparametro_gamma.png")
```

dove `Γ(k, θ)` rappresenta una distribuzione gamma con parametro di forma `k` e parametro di scala `θ`

Questo può essere fatto con l'oggetto `BayesianRidge` (parte del modulo `linear_model`). Come tutti gli oggetti precedenti di regressione, questo può essere inizializzato senza argomenti richiesti.

### Time to Code!
Si usa l'oggetto `LogisticRegression`del modulo `linear_model` per la classificazione multiclasse.
- La funzione `multiclass_lr` adatterà  um moddello di regressione logistica a un set du dati con etichette multiclasse.
- Setta `reg` uguale a `linear_model.LogisticRegression` con le kewyword `solver` e `max_iter` con argomenti `lbfgs` e `max_iter`. Ha anche la keyword `multi_class` con argomento `multinomial`.
- Chiama `reg.fit` con `data` e `labels` come 2 argomenti di input. Si ritorni `reg`.

```python
def multiclass_lr(data, labels, max_iter):
    reg = linear_model.LogisticRegression(solver='lbfgs', 
                                          max_iter=max_iter, 
                                          multi_class='multinomial')
    reg.fit(data, labels)
    return reg
```

## Decision Trees
### Making decision
Ogni modello che è stato esaminato, si basa sulla creazione di una combinazione lineare ottimale di funzionalità del set di dati per la **regressione** la **classificazione**.
- Un altro modello popolare in data scientist sia per la classificazione, sia per la regressione è il `decision tree`. Si tratta di un `binary tree` in cui ogni nodo dell'albero decide su una particolare funzionalità del set di dati e si scende al figlio sinistro o destro a seconda del valore della funzionalità.

Se una feature è **booleana**, si va a sinistra o a destra dal nodo a seconda che il valore della feature sia `True` o `False`. Se una feature è numerica, si può decidere di andare a sinistra o a destra in base a un limite di decisione per la feature.
<br><br>Le foglie dell'albero determinano l'etichettaa di classe da prevedere (in classificazione) o il valore minimo reale da prevedere in (in regressione).
<br><br>Esempio di `classificazione multiclasse`.

```python
Image("classificazione_multiclasse.png")
```

<br><br>Il codice sottostante dimostra come creare alberi di decisione per **decisione** e **regressione**: Ogni albero di decisione usa la funzione `fit` per adattare i dati e le etichette.

### Choosing features
Poichè un albero decisionale prende decisioni basate sui valori delle funzionalità, la domanda ora diventa come si scelgano le funzionalità sui cui decidere ogni nodo. In particolare:

- per gli alberi di `regressione` si sceglie la funzionalità in ogni nodo che minimizzi l'`MSE` o il `MAE` delle restanti osservazioni del set di dati, a seconda di quale si sceglie di utilizzare (L'impostazione predefinita per `DecisionTreeRegressor` è MSE).

## Training and Testing
### Training and testing sets
Si è parlato di come adattare un modello su dati ed etichette. TUttavia una volta adattato il modello come lo si valuta?
- è una cattiva idea valutare un modello esclusivamente sullo stesso set di dati su cui è stato inserito, perchè i parametri del modello sono già sintonizzati per quel set di dati.
- invece si può dividere il set di dati originale in due set di dati: uno per il `training` e uno per il `testing`.

Il set di training è usato per adattare il modello ai dati, mentre il set di testing è usato per valutare il modello. Tuttavia poichè ciò riducele dimensioni del set di test, vi è una maggior probabilità che il set di test non sia rappresentativo del set di dati originale, il che può portare a una valutazione imprecisa.
<br><br> In generale, il set di **testing** è circa il `10-30%` del set di dati originale, mentre il set di **training** costituisce il restante `70-90%`.

### Time to Code!
L'esercizio per questo capitolo, sarà per finire un utile funzione chiamata `dataset_splitter`, che sarà usata nei capitoli a seguire.
<br><br>La funzione dividerà i dataset di input di `training` e `testing` e poi raggruppare i dati e le etichette in base al tipo di set.
<br>
- Setta `split_dataset` uguale a `train_test_split` applicati a `data` e `labels` come argomenti richiesti, così come `test_size` per l'argomento della keyword `test_size`.
- Setta `train_set` uguale a una tupla contenente il primo e il terzo elemento di `split_test.
- Setta anche `test_set` uguale a una tupla contenente il secondo e quarto elemento di `split_test`.
- Ritorna una tupla contenente, in questo ordine, `train_set` e `test_data`.

```python
def dataset_splitter(data, labels, test_size=0.25):
    split_dataset = train_test_split(data, 
                                     labels,
                                     test_size=test_size)
    
    train_set = (split_dataset[0], 
                 split_dataset[2])
    
    test_set = (split_dataset[1], 
                split_dataset[3])
    
    return train_set, test_set
```

## Cross-Validation
### Additional evaluation datasets
A volte, non è sufficiente avere un singolo set di testing per la valutazione del modello. Avere set aggiuntivi di dati per la valutazione dà una misurazione più accurata di quanto sia buono il modello per il set di dati originale.
<br><br>Se il set di dati orginale è abbastanza grande lo si può effettivamente dividere in tre sottoinsiemi: `training`, `testing` e `validation`. Il set di validation ha le stesse dimensioni del set di testing ed è usato per valutare il modello dopo il training. Il set di testing è poi usato per una valutazione finale, una volta che il modello è stato allentato e ha superato la fase di tuning.
<br><br>
Tuttavia il partizionamento del set di dati orginale in tre set distinti ridurrà le dimensioni del set di training. Ciò può ridurre le prestazioni del modello se il set di dati originale non è abbastanza grande. Una soluzione a questo problema è la `cross-validation (CV)`

```python
Image("cross-validation.png")
```

Ad ogni round del `K-Fold algorithm`, il modello viene addestrato sul set di training di quel round  (the combined training folds) e poi viene valutato sul singolo validation fold. 
<br>La metrica di valutazione dipende dal modello. 
- Per i modelli di classificazione, si tratta solitamente dell'accuratezza della `classificazione` sul set di convalida. 
- Per i modelli di `regressione`, questo può essere l'errore quadratico medio del modello, l'errore assoluto medio o il valore R2 sul set di convalida.

### Time to Code!
L'esercizio di codifica per questo capitolo consiste nel completare la summenzionata funzione `cv_decision_tree`.
- Il primo argomento della funzione definisce se l'albero decisionale è per la classificazione/regressione, 
- i due argomenti successivi rappresentano i dati/le etichette 
- e gli ultimi due argomenti rappresentano rispettivamente la profondità massima e il numero di pieghe dell'albero.

Per prima cosa si crei il decisio tree (usando il modulo `tree`).
- Si inizializzi `d_tree` con `tree.DecisionTreeClassifier` se `is_clf` è `True`, altrimenti si usi `tree.DecisionTreeRegressor`. In entrambi i casi si inizializzi l'argomento della keyword `max_depth` con `max_depth`.
- Poi si usi la funzione `cross_val_score` per ottenere i punteggi di CV.
- Setta `scores` uguale a `cross_val_score` applicato con `d_tree`, `data`, e `labels` per i primi 3 argomenti. Usa `cv=cv`, poi ritorna `scores`.

```python
def cv_decision_tree(is_clf, data, labels, max_depth, cv):
    if is_clf:
        d_tree = tree.DecisionTreeClassifier(max_depth=max_depth)
    else:
        d_tree = tree.DecisionTreeRegressor(max_depth=max_depth)
        scores = cross_val_score(d_tree, 
                                 data, 
                                 labels, cv=cv)
    return scores
```
